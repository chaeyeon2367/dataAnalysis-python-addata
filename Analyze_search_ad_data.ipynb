# Preprocessing data based on advertising groups

```python
#Sum of group data - Impression, clicks, total cost column
grouped.sum()
```

```python
df_group=grouped.sum()
df_group
# Pre-processing click-rate, average click-cost column
# As-is : Sum of click rate for keyword, sum of average click cost for keyword
# to-be : Click rate for advertising group, average click cost for advertising group
```

```python
#Click Rate (ctr) = Clicks / Impression Count
#Data preprocessing - numerical computation in columns of data frames
df_group['클릭률(%)']=df_group['클릭수']/df_group['노출수']
```

```python
#Average click cost (cpc) = total cost / number of clicks
df_group['평균클릭비용(VAT포함,원)'] = df_group['총비용(VAT포함,원)']/df_group['클릭수']
```

```python
#Data with 0 in the number of clicks column exists => Missing value in the average click cost column exists
df_group[df_group['클릭수']==0]
```

```python
#fillna - function that replaces missing values with passed values
#Average click cost
df_group['평균클릭비용(VAT포함,원)']=df_group['평균클릭비용(VAT포함,원)'].fillna(0)
```

```python
df_group[df_group['클릭수']==0]
```

```python
# Average click cost column rounding (round), decimal point elimination (astype(int)
df_group['평균클릭비용(VAT포함,원)'] = round(df_group['평균클릭비용(VAT포함,원)'],0)
df_group['평균클릭비용(VAT포함,원)'] = df_group['평균클릭비용(VAT포함,원)'].astype(int)
```

### # Focused management advertising group / Low efficiency advertising group

- Focused management advertising group
1. Top 20% (over 80 percentile of exposure)
2. Top 10% (90% or more clicks)

```python
imp=df_group['노출수']
clk=df_group['클릭수']
```

```python
result = df_group[(imp>=imp.quantile(0.8))&(clk>=clk.quantile(0.9))]
```

```python
#index - Focused Management Advertising Group Output
result.index
```

- Low efficiency advertising group:
Advertising groups that are expensive even though they are not the top keywords based on exposure and clicks.
1. Impression less than 80 percentile
2. Less than 90 percent of clicks
3. Total cost 60 percent or higher
4. Total cost less than 90 percentile

```python
cost = df_group['총비용(VAT포함,원)']
```

```python
result=df_group[(imp<imp.quantile(0.8))&(clk<clk.quantile(0.9))&(cost>=cost.quantile(0.6))&(cost<cost.quantile(0.9))]
```

```python
#index - Low efficiency advertising group output
result.index
```
